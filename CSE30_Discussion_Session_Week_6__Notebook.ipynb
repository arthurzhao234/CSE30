````python
""" Write Code Here """

from datetime import date
from functools import reduce

# Step 1: Get input and convert to list of integers (one-liner using map)
numbers = list(map(int, input().split()))

# Step 2: Convert every group of 3 numbers into a date object using a lambda function
dates = list(map(lambda i: date(*numbers[3*i:i*3+3]), range(len(numbers)//3)))

# Step 3: Filter dates that fall in leap years and sort them
leap_dates = list(filter(lambda d: d.year % 4 == 0, sorted(dates)))

# Step 4: Use reduce to sum periods (in years) between consecutive leap years
periods = [leap_dates[i].year - leap_dates[i-1].year for i in range(1, len(leap_dates))]
total_period=reduce(lambda x,y:x+y,periods)

# Output
print("Leap year dates (sorted):")
for d in leap_dates:
    print(d.isoformat())

print("Total period between leap years:", total_period)
````

### **4. Buttons and Click Events**

The following code illustrates how to simulate event-driven programming in Google Colab. Although Colab is not inherently suited for such paradigms, the `ipywidgets` module provides a partial workaround.

This code creates a button in a Colab notebook and attaches an event handler that runs a function when the button is clicked.
Let's go over it step-by-step:

* First, we import the `ipywidgets` library for interactive widgets (like a button).
    ```
    import ipywidgets as widgets
    from IPython.display import display
    ```
* We creates a button widget with the label "Click Me".

    ```
    button = widgets.Button(description="Click Me")
    ```

* We define a function (event-handler) that runs when the button is clicked. The parameter `b` represents the button object (required by the event system).

    ```
    def on_button_click(b):
        print("Button was clicked!")
    ```

* We bind the button (listener) to the event-handler, so the function is triggered by user interaction. In other words, we registers the `on_button_click` function as an event handler for the button.

    The button acts as the event source, and the `on_click` method binds a click event to the event handler, which defines the response when the button is clicked.

    ```
    button.on_click(on_button_click)
    ```
* We use the `display` function to render the button in the notebook so the user can interact with it.
    ```
    display(button)
    ```


````python
import ipywidgets as widgets
from IPython.display import display

# Create a button
button = widgets.Button(description="Click Me")

# Define an event handler
def on_button_click(b):
    print("Button was clicked!")

# Attach the event
button.on_click(on_button_click)

# Display the button
display(button)
````

Write code that creates a Tic-Tac-Toe board using buttons, where each button toggles between $X$ and $O$ when clicked. There are multiple ways to implement this -- just ensure that the buttons respond to clicks.

The starter code is provided below. You can fill in the missing parts.

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

""" Write Code Here """

# Initialize game state
current_player = ['X']  # Using list for mutability inside handler
board = [''] * 9        # 3x3 board

# Create 9 buttons
buttons = [widgets.Button(description='', layout=widgets.Layout(width='60px', height='60px')) for _ in range(9)]

# Define click handler
def on_click(i):
    def handle_click(b):
        if b.description == '':  # modifies the button label if the button is not labeled
            b.description = current_player[0]
            board[i] = current_player[0]
            if current_player[0]=='0':
              current_player[0] = 'X'
            else:
              current_player[0] = '0'   # alternates X and O
    return handle_click

# Attach handlers
for i, btn in enumerate(buttons):
    btn.on_click(on_click(i))

# Display as 3x3 grid
grid = widgets.GridBox(buttons, layout=widgets.Layout(
    grid_template_columns="repeat(3, 60px)",
    grid_gap="5px"
))
display(grid)
````
